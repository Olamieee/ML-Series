```python
df = pd.DataFrame(digits.data, columns= digits.feature_names)
df.head()
```

```python
df['target'] = digits.target
df.head()
```

```python
df[df.target==1].head()
```

```python
df[df.target==8].head()
```

```python
df['name'] = df.target.apply([lambda x: digits.target_names[x]])
df.head()
```

```python
df0 = df[df.target==0] 
df1 = df[df.target==1] 
df2 = df[df.target==2] 
df3 = df[df.target==3] 
df4 = df[df.target==4] 
df5 = df[df.target==5] 
df6 = df[df.target==6] 
df7 = df[df.target==7] 
 
```

```python
x = df.drop(['target', 'name'], axis='columns')
x.head()
```

```python
y = df.target
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.8)
len(x_train)
x_train
```

```python
len(x_test)
x_test
```

```python
from sklearn.svm import SVC
model = SVC()
model.fit(x_train, y_train)
```

```python
model.score(x_test,y_test)
```

**Tune Parameters**

**1. Regularization (C)**

```python
model_c = SVC(C=10)
model_c.fit(x_train, y_train)
```

```python
model_c.score(x_test,y_test)
```

```python
model_c.score(x_train, y_train)
```

**2. Gamma**


```python
model_g = SVC(gamma=10)
model_g.fit(x_train, y_train)
```

```python
model_g.score(x_train,y_train)
```

```python
model_g.score(x_test,y_test)
```

**3. Kernel**

```python
model_k = SVC(kernel='rbf')
model_k.fit(x_train, y_train)
```

```python
model_k.score(x_test,y_test)
```

```python
model_k.score(x_train,y_train)
```

```python
model_k = SVC(kernel='linear')
model_k.fit(x_train, y_train)
```

```python
model_k.score(x_test,y_test)
```

```python
model_k.score(x_train,y_train)
```
